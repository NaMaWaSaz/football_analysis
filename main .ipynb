# 📦 Import required modules

```python
from utils import read_video, save_video
from trackers import Tracker
import numpy as np
from team_assigner import TeamAssigner
from player_ball_assigner import PlayerBallAssigner
from camera_movement_estimator import CameraMovementEstimator
from view_transformer import ViewTransformer
from speed_and_distance_estimator import SpeedAndDistance_Estimator
from tqdm import tqdm
import os
```

#Main function

```python
def main():
    MAX_FRAMES = 300  # for quick tests; set to 0 or None to disable

    # 1. Read video
    print("📥 Reading video…")
    video_frames = read_video('football_analysis/Copie de real.mp4')
    total = len(video_frames)
    print(f"   → {total} frames loaded")
    if MAX_FRAMES and total > MAX_FRAMES:
        print(f"⚠️ Debug: limiting to first {MAX_FRAMES} frames")
        video_frames = video_frames[:MAX_FRAMES]
        total = MAX_FRAMES

    # 2. Tracker + position stubs
    print("🤖 Initializing tracker & loading tracks…")
    tracker = Tracker('models/best.pt')
    tracks = tracker.get_object_tracks(
        video_frames,
        read_from_stub=True,
        stub_path='stubs/track_stubs.pkl'
    )
    tracker.add_position_to_tracks(tracks)
    print(f"   → Tracks ready for {len(tracks['players'])} frames")

    # 3. Camera movement
    print("🎥 Estimating camera movement…")
    cam_est = CameraMovementEstimator(video_frames[0])
    cam_movements = cam_est.get_camera_movement(
        video_frames,
        read_from_stub=True,
        stub_path='stubs/camera_movement_stub.pkl'
    )
    cam_est.add_adjust_positions_to_tracks(tracks, cam_movements)
    print(f"   → Camera movement for {len(cam_movements)} frames applied")

    # 4. View transform
    print("🔄 Applying view transformation…")
    vt = ViewTransformer()
    vt.add_transformed_position_to_tracks(tracks)

    # 5. Ball interpolation
    print("🏀 Interpolating ball positions…")
    tracks["ball"] = tracker.interpolate_ball_positions(tracks["ball"])

    # 6. Speed & distance
    print("🚀 Calculating speed & distance…")
    sd_est = SpeedAndDistance_Estimator()
    sd_est.add_speed_and_distance_to_tracks(tracks)

    # 7. Team assignment
    print("🏷️  Assigning teams & colors…")
    ta = TeamAssigner()
    ta.assign_team_color(video_frames[0], tracks['players'][0] if tracks['players'] else {})
    for i, ptrack in enumerate(tracks['players']):
        for pid, info in ptrack.items():
            bbox = info.get('bbox')
            if bbox is None:
                print(f"   ⚠️ Frame {i}, player {pid}: missing bbox, skipping")
                continue
            try:
                team = ta.get_player_team(video_frames[i], bbox, pid)
                info['team'] = team
                info['team_color'] = ta.team_colors.get(team)
            except Exception as e:
                print(f"   ❌ Frame {i}, player {pid}: get_player_team error: {e}")
    print("   → Team assignment done")

    # 8. Ball possession
    print("⚽ Determining ball possession…")
    pa = PlayerBallAssigner()
    ball_ctrl = []
    for i, ptrack in enumerate(tracks['players']):
        ball_info = tracks['ball'][i].get(1, {})
        bbox = ball_info.get('bbox')
        if bbox is None:
            ball_ctrl.append(ball_ctrl[-1] if ball_ctrl else None)
            continue
        try:
            ap = pa.assign_ball_to_player(ptrack, bbox)
            if ap != -1:
                tracks['players'][i][ap]['has_ball'] = True
                ball_ctrl.append(tracks['players'][i][ap].get('team'))
            else:
                ball_ctrl.append(ball_ctrl[-1] if ball_ctrl else None)
        except Exception as e:
            print(f"   ❌ Frame {i}: ball assign error: {e}")
            ball_ctrl.append(ball_ctrl[-1] if ball_ctrl else None)
    ball_ctrl = np.array(ball_ctrl)
    print(f"   → Ball control array length {len(ball_ctrl)}")

    # 9. Draw annotations
    print("🎨 Drawing annotations…")
    annotated = tracker.draw_annotations(video_frames, tracks, ball_ctrl)

    # 10. Overlay camera movement with progress
    print("📷 Overlaying camera movement…")
    overlaid = []
    for frame, mov in tqdm(zip(annotated, cam_movements),
                           total=len(annotated),
                           desc="Camera Overlay"):
        try:
            out = cam_est.draw_camera_movement([frame], [mov])[0]
            overlaid.append(out)
        except Exception as e:
            print(f"   ❌ Overlay error on frame: {e}")
            overlaid.append(frame)
    annotated = overlaid

    # 11. Overlay speed & distance
    print("📏 Overlaying speed & distance…")
    try:
        sd_est.draw_speed_and_distance(annotated, tracks)
    except Exception as e:
        print(f"   ❌ Speed/distance overlay error: {e}")

    # 12. Save final video
    os.makedirs('output_videos', exist_ok=True)
    out_path = 'output_videos/output_jemni.avi'
    print(f"💾 Saving to {out_path}…")
    save_video(annotated, out_path)
    print("✅ Video saved successfully!")

if __name__ == '__main__':
    main()
```

```python
import os
import cv2
import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi

# Paths
clip_dir = 'output_videos'
track_stub_path = 'stubs/track_stubs.pkl'

# 1. Find video clip
clips = [f for f in os.listdir(clip_dir) if f.lower().endswith(('.mp4','.avi','.mov'))]
if not clips:
    raise FileNotFoundError(f"No video clips in {clip_dir}")
video_path = os.path.join(clip_dir, clips[0])
print("Using clip:", video_path)

# 2. Read first frame
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
cap.release()
if not ret:
    raise IOError("Cannot read first frame")
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # for matplotlib

# 3. Load tracks
with open(track_stub_path, 'rb') as f:
    tracks = pickle.load(f)
print(f"Loaded tracks for {len(tracks['players'])} frames. Frame shape: {frame_rgb.shape}")
```

#Voronoi Diagram

```python
# Choose your frame index
frame_idx = 0

# Extract bboxes → centers
frame_tracks = tracks['players'][frame_idx]
points = []
for info in frame_tracks.values():
    bbox = info.get('bbox')
    if bbox:
        x1, y1, x2, y2 = bbox
        cx = (x1 + x2) / 2
        cy = (y1 + y2) / 2
        points.append((cx, cy))
points = np.array(points)

print(f"Frame {frame_idx}: extracted {len(points)} player points")
```

```python
# Compute raw Voronoi
vor = Voronoi(points)

# We’ll build finite polygons by clipping each region to the image rectangle
# Reference: https://stackoverflow.com/a/20678647

from shapely.geometry import Polygon
from shapely.ops import clip_by_rect

h, w, _ = frame_rgb.shape
bbox = Polygon([(0,0), (w,0), (w,h), (0,h)])

regions = []
for region_idx in vor.point_region:
    vert_ids = vor.regions[region_idx]
    if not vert_ids or -1 in vert_ids:
        # skip infinite regions
        regions.append(None)
        continue
    poly = Polygon([vor.vertices[i] for i in vert_ids])
    # clip to frame bounds
    clipped = poly.intersection(bbox)
    regions.append(clipped)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.imshow(frame_rgb)

# draw each cell
for poly, pt in zip(regions, points):
    if poly and not poly.is_empty:
        xs, ys = poly.exterior.xy
        ax.fill(xs, ys, alpha=0.4, edgecolor='white', linewidth=2)
# draw the seed points
ax.scatter(points[:,0], points[:,1], c='yellow', s=80, edgecolor='black', linewidth=1)

ax.set_title(f'Voronoi Diagram – Frame {frame_idx}', fontsize=18)
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# --- essentials
import os, pickle, cv2
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# (networkx is pre-installed in Colab; if not, uncomment)
# !pip install networkx

plt.rcParams["figure.facecolor"] = "#1b5e20"   # dark-green canvas
```

```python
# pick any frame to inspect
idx = 0
print("Sample players dict  ➜  tracks['players'][idx]")
print(tracks['players'][idx])          # players dict {pid: info}

print("\nSample ball dict     ➜  tracks['ball'][idx]")
print(tracks['ball'][idx])             # ball detections (likely key 1)
```

```python
def centre(b):
    """bbox -> (cx, cy)"""
    x1, y1, x2, y2 = b
    return ( (x1+x2)/2, (y1+y2)/2 )

# pre-compute ball centre each frame  (uses first entry in the ball dict)
ball_xy = []
for bdict in tracks['ball']:
    if bdict:                            # some frames might be empty
        bx = next(iter(bdict.values()))  # first detection
        ball_xy.append( centre(bx['bbox']) )
    else:
        ball_xy.append( None )
```

```python
pass_events   = []
prev_owner    = None
frames_gap    = 3        # ignore flickers (owner must persist ≥3 frames)
min_transfer  = 100      # px ball must travel between owners
max_radius_px = 60       # player must be within this px of ball to own

last_change_f = -999     # frame index of last owner change

for f, ptrack in enumerate(tracks['players']):
    bxy = ball_xy[f]
    if bxy is None:            # ball missing
        continue

    # ----- find nearest player -----
    owner, d_min = None, max_radius_px
    for pid, info in ptrack.items():
        px, py = centre(info['bbox'])
        d      = np.hypot(px-bxy[0], py-bxy[1])
        if d < d_min:
            owner, d_min = pid, d

    # ----- pass detection -----
    if owner is not None and owner != prev_owner:
        # Must satisfy: enough frames gap + ball moved far
        if (f - last_change_f) >= frames_gap:
            moved = ( np.hypot(*(np.array(ball_xy[f]) - np.array(ball_xy[last_change_f]))
                      ) if prev_owner is not None else min_transfer )
            if moved >= min_transfer:
                pass_events.append( (prev_owner, owner) )
                last_change_f = f
        prev_owner = owner

print("Recovered", len(pass_events), "passes")
```

```python
# ---------- graph ----------
wgt = {}
for u, v in pass_events:
    if u is None or v is None: continue
    wgt[(u, v)] = wgt.get((u, v), 0) + 1

G = nx.DiGraph()
for pid, pos in tracks['players'][0].items():    # use frame-0 for node set
    G.add_node(pid, pos=centre(pos['bbox']))

for (u, v), wt in wgt.items():
    if (u in G) and (v in G):
        G.add_edge(u, v, weight=wt)

# ---------- visual ----------
fig, ax = plt.subplots(figsize=(13, 8))
ax.imshow(pitch_rgb)

for u, v, data in G.edges(data=True):
    x1, y1 = G.nodes[u]['pos']
    x2, y2 = G.nodes[v]['pos']
    wt     = data['weight']
    ax.annotate('', xy=(x2,y2), xytext=(x1,y1),
                arrowprops=dict(arrowstyle='-|>', color='white',
                                linewidth=0.8+0.4*wt,
                                alpha=min(0.2+0.08*wt,0.9)))

for pid, info in G.nodes(data=True):
    x, y = info['pos']
    deg  = G.degree(pid, weight='weight')
    ax.scatter(x, y, s=300+20*deg, c='#ffeb3b',
               edgecolor='black', linewidth=2, zorder=3)
    ax.text(x, y, str(pid), ha='center', va='center',
            fontsize=10, fontweight='bold', color='black', zorder=4)

ax.set_title('Pass Network', fontsize=22, color='white')
ax.axis('off')
plt.tight_layout()
plt.show()
```
